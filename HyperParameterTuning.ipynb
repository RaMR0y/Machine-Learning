## Why hyper-parameter tuning?

Hyperparameter tuning is crucial for several reasons:

### 1. **Performance Optimization**:
   - **Accuracy and Generalization**: Proper tuning of hyperparameters can significantly improve the accuracy and generalization of a model. Different combinations of hyperparameters can lead to vastly different performance levels.
   - **Model Efficiency**: Efficient hyperparameter settings can make the training process faster and more efficient, reducing computational costs and time.

### 2. **Model Robustness**:
   - **Stability**: Hyperparameter tuning can help in finding a stable model that performs well across different datasets and scenarios, making the model more robust.
   - **Avoid Overfitting/Underfitting**: Proper tuning can help balance the trade-off between overfitting and underfitting, ensuring the model is neither too complex nor too simple.

### 3. **Resource Management**:
   - **Computational Resources**: Optimal hyperparameter settings can make better use of available computational resources (e.g., GPU memory, processing power), preventing wastage and avoiding out-of-memory errors.
   - **Time Management**: Efficient models reduce training time, which is crucial in scenarios where quick iteration and deployment are necessary.

### 4. **Understanding Model Behavior**:
   - **Insights**: Tuning hyperparameters provides insights into how different aspects of the model affect its performance. This understanding can guide future model development and adjustments.
   - **Parameter Sensitivity**: Understanding the sensitivity of the model to various hyperparameters helps in diagnosing issues and making informed decisions about model architecture and training strategies.

### 5. **Competitiveness**:
   - **Benchmarking**: In competitive fields such as machine learning competitions (e.g., Kaggle), fine-tuned hyperparameters can be the difference between top-ranking models and average ones.
   - **State-of-the-Art Performance**: Achieving state-of-the-art performance on benchmarks and datasets often requires meticulous hyperparameter tuning.

### 6. **Adaptability**:
   - **Transfer Learning**: When transferring a model to a new task or domain, tuning hyperparameters can help adapt the model more effectively to the new data, improving performance.
   - **Scalability**: As models scale up (e.g., deeper networks, larger datasets), tuning becomes even more critical to manage complexity and maintain performance.

### 7. **Optimization of Trade-offs**:
   - **Speed vs. Accuracy**: In many real-world applications, there is a trade-off between speed and accuracy. Hyperparameter tuning helps find the right balance based on the specific requirements of the application.

### Examples of Hyperparameters:
- **Learning Rate**: Controls how much to change the model in response to the estimated error each time the model weights are updated.
- **Batch Size**: Determines the number of samples that will be propagated through the network.
- **Dropout Rate**: Prevents overfitting by randomly setting a fraction of input units to 0 at each update during training time.
- **Number of Hidden Units**: Influences the model's capacity and its ability to learn complex patterns.

By carefully tuning these and other hyperparameters, practitioners can significantly improve the performance, efficiency, and robustness of machine learning models, leading to better outcomes in various applications.

### Best Practices for Hyperparameter Tuning

Hyperparameter tuning can be a complex and resource-intensive process. Here are some best practices and guidelines to follow:

#### 1. **Use a Representative Subset of Data**
- **Portion of Data**: Initially, use a subset of your dataset to speed up the tuning process. This subset should be representative of the entire dataset. Typically, using 10-30% of the data is a good starting point.
- **Full Dataset**: Once you find promising hyperparameter configurations, you can refine these settings using the full dataset for final training.

#### 2. **Monitor Relevant Metrics**
- **Loss Metrics**: Monitor both training and validation loss to ensure the model is learning and generalizing well.
- **Accuracy and Other Metrics**: Depending on the problem, monitor accuracy, precision, recall, F1-score, etc., on both training and validation sets to get a holistic view of model performance.
- **Early Stopping**: Use early stopping to terminate training if the model stops improving on the validation set, which helps prevent overfitting and saves computational resources.

#### 3. **Cross-Validation**
- **Cross-Validation**: Use k-fold cross-validation to ensure that your hyperparameter tuning results are robust and not due to random chance. It helps in assessing the model’s performance more reliably.
- **Stratified K-Folds**: For classification problems, use stratified k-folds to maintain the same class distribution in each fold as in the entire dataset.

#### 4. **Search Strategies**
- **Random Search**: Randomly sample hyperparameter values. It's often more efficient than grid search, especially when dealing with a large number of hyperparameters.
- **Bayesian Optimization**: Use Bayesian optimization techniques (e.g., Optuna) to explore the hyperparameter space intelligently and find optimal configurations more efficiently.
- **Grid Search**: Use grid search for smaller hyperparameter spaces or when you want to ensure exhaustive exploration of specific ranges.

#### 5. **Incremental Refinement**
- **Coarse to Fine**: Start with a coarse search over a broad range of hyperparameters. Narrow down the ranges and refine the search based on initial results.
- **Hierarchical Tuning**: Tune important hyperparameters first. Once you have a good configuration, fix those and tune less critical hyperparameters.

#### 6. **Resource Management**
- **Parallelization**: Leverage parallel processing to run multiple hyperparameter configurations simultaneously, especially when using methods like random search or Bayesian optimization.
- **Hardware Utilization**: Ensure efficient use of hardware resources (e.g., GPUs, TPUs). Monitor resource utilization to avoid bottlenecks.

#### 7. **Logging and Reproducibility**
- **Experiment Tracking**: Use tools like TensorBoard, MLflow, or Optuna’s logging to track experiments, configurations, and results systematically.
- **Reproducibility**: Set random seeds and document the exact hyperparameter configurations and data preprocessing steps to ensure reproducibility.

#### 8. **Domain-Specific Considerations**
- **Domain Knowledge**: Incorporate domain knowledge to set sensible ranges and distributions for hyperparameters.
- **Custom Metrics**: Sometimes domain-specific custom metrics might provide better insights into model performance than standard metrics.

#### Pro Tips and Guidelines:

1. **Learning Rate Schedulers**: Tune learning rate schedules and decays as they can significantly impact convergence and performance.
2. **Warm-Up Phases**: For deep learning models, use learning rate warm-up strategies to stabilize initial training.
3. **Batch Size Impact**: Be aware of the impact of batch size on learning dynamics and computational efficiency. Smaller batch sizes offer more stochasticity but can slow down training.
4. **Regularization Techniques**: Regularization parameters (e.g., dropout rate, L2 regularization) are crucial to prevent overfitting and should be tuned carefully.
5. **Hyperparameter Sensitivity**: Identify and focus on tuning hyperparameters that your model is most sensitive to, as they can have a more significant impact on performance.
6. **Adaptive Methods**: Use adaptive optimization methods like Adam, RMSprop, and their variants. Tune their hyperparameters like learning rate and epsilon.
7. **Use Libraries**: Utilize hyperparameter tuning libraries like Optuna, Hyperopt, Ray Tune, and others to automate and manage the tuning process efficiently.
8. **Scaling Up**: Once you have an optimal set of hyperparameters on a smaller subset, scale up the tuning process using the full dataset and more compute resources.

### How to Create Your Own Hyperparameter Tuning Framework

Creating your own hyperparameter tuning framework involves several steps:

#### 1. Define the Search Space
- **Parameter Types**: Decide on the types of parameters (e.g., continuous, discrete, categorical).
- **Range and Distribution**: Specify the ranges and distributions for each parameter.

```python
search_space = {
    'learning_rate': {'type': 'continuous', 'range': [0.0001, 0.1], 'distribution': 'loguniform'},
    'batch_size': {'type': 'discrete', 'values': [16, 32, 64, 128]},
    'dropout_rate': {'type': 'continuous', 'range': [0.0, 0.5], 'distribution': 'uniform'}
}
```

#### 2. Implement Search Algorithms
- **Random Search**: Randomly sample hyperparameters from the search space.
- **Grid Search**: Exhaustively search over a predefined grid of hyperparameters.
- **Bayesian Optimization**: Use a probabilistic model to predict the performance of hyperparameter configurations and select the most promising ones.

```python
def random_search(search_space, num_trials):
    trials = []
    for _ in range(num_trials):
        trial = {param: np.random.uniform(*search_space[param]['range']) if search_space[param]['type'] == 'continuous' else np.random.choice(search_space[param]['values']) for param in search_space}
        trials.append(trial)
    return trials
```

````python
import torch
import torch.nn.functional as F
import torchvision.transforms as transforms
from torchvision.datasets import FashionMNIST
from torch.utils.data import DataLoader, random_split
import pytorch_lightning as pl
from pytorch_lightning import LightningModule, Trainer
from pytorch_lightning.tuner.tuning import Tuner
from pytorch_lightning.loggers import TensorBoardLogger
import optuna
import logging

# Silence PyTorch Lightning INFO messages
logging.getLogger("pytorch_lightning").setLevel(logging.WARNING)

class FashionMNISTModel(LightningModule):
    def __init__(self, lr=1e-3, batch_size=64, dropout=0.5, hidden_units=128):
        super(FashionMNISTModel, self).__init__()
        self.lr = lr
        self.batch_size = batch_size
        self.layer_1 = torch.nn.Linear(28 * 28, hidden_units)
        self.layer_2 = torch.nn.Linear(hidden_units, hidden_units // 2)
        self.layer_3 = torch.nn.Linear(hidden_units // 2, 10)
        self.dropout = torch.nn.Dropout(dropout)

    def forward(self, x):
        x = x.view(x.size(0), -1)
        x = F.relu(self.layer_1(x))
        x = self.dropout(x)
        x = F.relu(self.layer_2(x))
        x = self.dropout(x)
        x = self.layer_3(x)
        return F.log_softmax(x, dim=1)

    def training_step(self, batch, batch_idx):
        x, y = batch
        y_hat = self(x)
        loss = F.nll_loss(y_hat, y)
        self.log('train_loss', loss)
        return loss

    def configure_optimizers(self):
        return torch.optim.Adam(self.parameters(), lr=self.lr)

    def train_dataloader(self):
        transform = transforms.Compose([transforms.ToTensor()])
        dataset = FashionMNIST(root=".", train=True, download=True, transform=transform)
        return DataLoader(dataset, batch_size=self.batch_size, num_workers=4, shuffle=True)

def objective(trial):
    # Hyperparameters to tune
    lr = trial.suggest_loguniform('lr', 1e-5, 1e-1)
    batch_size = trial.suggest_categorical('batch_size', [32, 64, 128])
    dropout = trial.suggest_uniform('dropout', 0.2, 0.5)
    hidden_units = trial.suggest_categorical('hidden_units', [64, 128, 256])

    model = FashionMNISTModel(lr=lr, batch_size=batch_size, dropout=dropout, hidden_units=hidden_units)

    trainer = Trainer(
        max_epochs=5,
        logger=TensorBoardLogger("logs", name="fashion_mnist"),
        limit_train_batches=0.1,  # Limit the training data to 10%
        accelerator='gpu' if torch.cuda.is_available() else 'cpu',
        devices=1 if torch.cuda.is_available() else None,
        enable_progress_bar=False,  # Suppress progress bar for cleaner output,
        log_every_n_steps = 10,
    )

    tuner = Tuner(trainer)
    tuner.scale_batch_size(model, max_trials=7)

    # Train the model
    trainer.fit(model)

    # Retrieve the final loss
    final_loss = trainer.callback_metrics["train_loss"].item()

    return final_loss

study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=50)

print(f"Best trial: {study.best_trial.value}")
print("Best hyperparameters: ", study.best_trial.params)
````

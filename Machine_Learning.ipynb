# What is machine learning?

Machine Learning (ML) is a subset of artificial intelligence (AI) that enables systems to learn and improve from experience without being explicitly programmed. In machine learning, algorithms use data to identify patterns and make decisions or predictions. Instead of following static instructions, these systems adapt and evolve based on the data they are exposed to.

Key components of ML include:

- **Data**: The foundational input used to train models. It can be in the form of numbers, images, text, or other types of information.
- **Models**: Mathematical constructs that define how the system will learn from the data and make predictions.
- **Training**: The process of feeding data to the model so it can adjust and improve its performance.
- **Inference**: The use of the trained model to make predictions or decisions on new data.

# ML Categorization

Machine Learning (ML) can be classified into different groups based on the nature of the learning process, the type of data, and the goals of the model. Here are the primary classifications:

### 1. **Supervised Learning**
   - **Definition**: In supervised learning, the model is trained using labeled data, where the input data and the corresponding correct output (target) are provided.
   - **Goal**: The model learns to map inputs to outputs, predicting the target for new, unseen data.
   - **Examples**:
     - Classification (e.g., spam detection, image recognition)
     - Regression (e.g., house price prediction, stock market forecasting)
   - **Algorithms**: Linear regression, Decision Trees, Support Vector Machines (SVM), Neural Networks

### 2. **Unsupervised Learning**
   - **Definition**: In unsupervised learning, the model is given data without labeled outputs. The model tries to find hidden patterns or intrinsic structures in the data.
   - **Goal**: To discover insights or groupings from the data, often without specific targets in mind.
   - **Examples**:
     - Clustering (e.g., customer segmentation, gene sequence analysis)
     - Dimensionality Reduction (e.g., Principal Component Analysis (PCA), t-SNE)
   - **Algorithms**: K-Means Clustering, Hierarchical Clustering, Autoencoders

### 3. **Semi-Supervised Learning**
   - **Definition**: This approach combines both labeled and unlabeled data. The model is trained on a small amount of labeled data and a large amount of unlabeled data.
   - **Goal**: To improve the learning accuracy by leveraging unlabeled data along with the labeled data.
   - **Examples**:
     - Image recognition tasks where labeling every image is costly, but some labels are available.
   - **Algorithms**: Self-training, Co-training, Generative Models

### 4. **Reinforcement Learning**
   - **Definition**: In reinforcement learning, an agent interacts with an environment and learns to make decisions by receiving rewards or penalties for actions.
   - **Goal**: To learn the best actions or policies to maximize cumulative rewards over time.
   - **Examples**:
     - Robotics (e.g., teaching a robot to walk)
     - Game AI (e.g., AlphaGo, autonomous gaming agents)
   - **Algorithms**: Q-Learning, Deep Q Networks (DQN), Proximal Policy Optimization (PPO)

### 5. **Self-Supervised Learning**
   - **Definition**: Self-supervised learning is a type of unsupervised learning where the system generates its own labels from the input data. The model is trained to predict part of the data from other parts.
   - **Goal**: To learn useful representations without relying on labeled data.
   - **Examples**:
     - Predicting missing parts of images, predicting the next word in a sentence.
   - **Algorithms**: Contrastive Learning, SimCLR, BERT (in NLP)

### 6. **Transfer Learning**
   - **Definition**: Transfer learning involves taking a pre-trained model (often from a large dataset) and fine-tuning it for a new, related task.
   - **Goal**: To leverage knowledge gained from one task to improve performance on a different but related task.
   - **Examples**:
     - Using a model trained on ImageNet for a medical image classification task.
   - **Algorithms**: Fine-tuning pre-trained Neural Networks

### 7. **Online Learning**
   - **Definition**: In online learning, the model is trained incrementally as new data comes in, rather than in batch mode.
   - **Goal**: To adapt to changes in data in real-time, making the system more flexible.
   - **Examples**:
     - Personalized recommendations that adjust based on recent user interactions.
   - **Algorithms**: Stochastic Gradient Descent (SGD), Passive-Aggressive Algorithms

### 8. **Multitask Learning**
   - **Definition**: Multitask learning involves training a model on multiple tasks simultaneously, sharing knowledge between tasks.
   - **Goal**: To improve generalization by learning shared representations for related tasks.
   - **Examples**:
     - Learning to classify images and predict image attributes (e.g., object detection and segmentation).
   - **Algorithms**: Neural Networks with shared layers, Multi-headed architectures

# LOGISTIC REGRESSION

### 1. **The Basic Idea:**
Imagine you're a doctor, and you're trying to predict whether a patient has a particular disease based on their age. You have data from other patients, and you know that older people tend to have a higher risk of the disease, but not every older person has it, and not every younger person is free of it. The prediction you make isn’t a simple “yes” or “no” — it’s about the **likelihood** of having the disease based on age.

### 2. **Why Not Just Draw a Straight Line?**
In **linear regression**, we could try to draw a straight line (like $ y = ax + b $) to predict the outcome. For example, we could try to say: "For every extra year of age, the likelihood of having the disease increases by some number." However, this wouldn't work well because probabilities must be between 0 and 1, and a straight line can extend beyond those limits (giving probabilities above 1 or below 0, which doesn't make sense).

### 3. **Logistic Regression to the Rescue:**
Logistic regression doesn’t predict “yes” or “no” directly. Instead, it predicts the **probability** that the outcome will be “yes” (the positive class), given certain input features (like age).

The key is that logistic regression doesn’t draw a straight line. Instead, it uses a **curve** called the **sigmoid function**, which squashes the values between 0 and 1. This way, no matter how high or low the input (age, in this case), the prediction will always be a value between 0 and 1 — a valid probability.

### 4. **The Sigmoid Function**:
The **sigmoid function** looks like an "S"-shaped curve:

$
\sigma(z) = \frac{1}{1 + e^{-z}}
$

- If $z$ is large and positive (e.g., the patient is very old, indicating high risk), the sigmoid function approaches 1, meaning a high probability that the person has the disease.
- If $z$ is large and negative (e.g., the patient is young, indicating low risk), the sigmoid function approaches 0, meaning a low probability of the disease.

This function ensures that the output stays within the range [0, 1], representing probabilities.

### 5. **What's Happening Behind the Scenes**:
Here’s the flow of how logistic regression works:
- **Step 1**: You take input data (like age) and multiply it by some weight $a$ (which represents the importance of age in predicting the disease).
- **Step 2**: You add a bias $b$, which shifts the curve left or right, depending on where the decision boundary is.
- **Step 3**: Instead of stopping here (which would be linear regression), you plug this linear result into the **sigmoid function**, which squashes the output between 0 and 1.
- **Step 4**: The result is the probability that the person has the disease.

So, if the sigmoid function outputs something like 0.8, it means there’s an 80% chance the person has the disease based on their age. If it outputs 0.2, it means there’s a 20% chance.

### 6. **The Decision Boundary**:
Once you have the probability, you can make a classification:
- If the probability is greater than 0.5, the model predicts the positive class (e.g., the person has the disease).
- If the probability is less than 0.5, the model predicts the negative class (e.g., the person does not have the disease).

In this sense, logistic regression is a way of drawing a **non-linear boundary** between the two classes (disease vs. no disease) using probabilities instead of strict lines.

### 7. **Intuitive Example: Deciding to Bring an Umbrella**
Imagine you’re trying to decide whether to bring an umbrella based on the weather. You look at various factors, like cloud cover, wind, and the weather forecast. Each of these factors contributes a certain "weight" to your decision.

- If it's very cloudy and the weather forecast says rain, the model will give a high probability (close to 1), and you’ll bring the umbrella.
- If the sky is clear and there's no wind, the model will give a low probability (close to 0), and you’ll decide not to bring the umbrella.

In both cases, the logistic regression model gives you a **probability** of rain, and based on that probability, you make your final decision.

### Summary of Key Points:
- **Logistic regression** predicts probabilities rather than raw outcomes.
- It uses the **sigmoid function** to squish the output between 0 and 1.
- It’s useful for classification tasks where you want to decide between two classes (yes/no, disease/no disease, rain/no rain).
- You can make a classification decision by setting a threshold (usually 0.5).

Logistic regression is intuitive because it models how we often think: “Given this information, what’s the chance that X will happen?” It’s a simple yet powerful tool for binary classification.

## Toy Example

**Manual computation of logistic regression** on a simple toy example, step by step, to understand how it works. We'll compute the predictions, loss (binary cross-entropy), gradients, and parameter updates using **gradient descent** for one iteration.

### 1. **Toy Example Setup**

We'll work with a single input feature $ x $ and one binary output $ y $. Here’s the data:

| $ x $  | $ y $ |
|:--------:|:-------:|
|   0.5    |    0    |
|   2.0    |    1    |
|   1.5    |    1    |
|   3.5    |    0    |

Our goal is to fit a logistic regression model where the predicted output is $ \hat{y} $, which represents the probability that $ y = 1 $. The equation of logistic regression is:

$
\hat{y} = \sigma(w \cdot x + b)
$

Where $ \sigma(z) = \frac{1}{1 + e^{-z}} $ is the **sigmoid function**.

We’ll start with initial parameters:
- **Weight $ w = 0.5 $**
- **Bias $ b = 0 $**
- **Learning rate $ \alpha = 0.1 $**

### 2. **Step-by-Step Computation**

#### Step 1: Forward Pass (Prediction)
For each data point, we calculate the **logistic regression prediction** $ \hat{y} $.

##### Data Point 1: $ x = 0.5 $, $ y = 0 $

1. Compute the linear combination (logit):
   $
   z = w \cdot x + b = 0.5 \cdot 0.5 + 0 = 0.25
   $
   
2. Apply the sigmoid function to get the predicted probability:
   $
   \hat{y} = \sigma(0.25) = \frac{1}{1 + e^{-0.25}} \approx 0.562
   $

##### Data Point 2: $ x = 2.0 $, $ y = 1 $

1. Compute the linear combination (logit):
   $
   z = 0.5 \cdot 2.0 + 0 = 1.0
   $
   
2. Apply the sigmoid function:
   $
   \hat{y} = \sigma(1.0) = \frac{1}{1 + e^{-1}} \approx 0.731
   $

##### Data Point 3: $ x = 1.5 $, $ y = 1 $

1. Compute the linear combination (logit):
   $
   z = 0.5 \cdot 1.5 + 0 = 0.75
   $
   
2. Apply the sigmoid function:
   $
   \hat{y} = \sigma(0.75) = \frac{1}{1 + e^{-0.75}} \approx 0.679
   $

##### Data Point 4: $ x = 3.5 $, $ y = 0 $

1. Compute the linear combination (logit):
   $
   z = 0.5 \cdot 3.5 + 0 = 1.75
   $
   
2. Apply the sigmoid function:
   $
   \hat{y} = \sigma(1.75) = \frac{1}{1 + e^{-1.75}} \approx 0.852
   $

#### Summary of Predictions:
- $ \hat{y}_1 \approx 0.562 $
- $ \hat{y}_2 \approx 0.731 $
- $ \hat{y}_3 \approx 0.679 $
- $ \hat{y}_4 \approx 0.852 $

#### Step 2: Compute the Loss (Binary Cross-Entropy)
For each data point, we compute the binary cross-entropy loss:

$
\text{Loss}(y, \hat{y}) = - \left[ y \log(\hat{y}) + (1 - y) \log(1 - \hat{y}) \right]
$

##### Data Point 1: $ y = 0 $, $ \hat{y} \approx 0.562 $

$
\text{Loss}_1 = - [0 \log(0.562) + (1 - 0) \log(1 - 0.562)] = - \log(1 - 0.562) \approx 0.825
$

##### Data Point 2: $ y = 1 $, $ \hat{y} \approx 0.731 $

$
\text{Loss}_2 = - [1 \log(0.731) + (1 - 1) \log(1 - 0.731)] = - \log(0.731) \approx 0.314
$

##### Data Point 3: $ y = 1 $, $ \hat{y} \approx 0.679 $

$
\text{Loss}_3 = - \log(0.679) \approx 0.387
$

##### Data Point 4: $ y = 0 $, $ \hat{y} \approx 0.852 $

$
\text{Loss}_4 = - \log(1 - 0.852) \approx 1.901
$

#### Average Loss:
$
\text{Average Loss} = \frac{1}{4} (0.825 + 0.314 + 0.387 + 1.901) \approx 0.857
$

#### Step 3: Compute the Gradients
Now, we calculate the gradient of the loss with respect to the weight $ w $ and the bias $ b $.

The gradient with respect to the weight $ w $ is:

$
\frac{\partial \text{Loss}}{\partial w} = \frac{1}{n} \sum_{i=1}^{n} (\hat{y}_i - y_i) x_i
$

The gradient with respect to the bias $ b $ is:

$
\frac{\partial \text{Loss}}{\partial b} = \frac{1}{n} \sum_{i=1}^{n} (\hat{y}_i - y_i)
$

##### Gradient for Weight $ w $:

$
\frac{\partial \text{Loss}}{\partial w} = \frac{1}{4} [(0.562 - 0) \cdot 0.5 + (0.731 - 1) \cdot 2.0 + (0.679 - 1) \cdot 1.5 + (0.852 - 0) \cdot 3.5]
$

$
\frac{\partial \text{Loss}}{\partial w} = \frac{1}{4} [0.281 + (-0.538) + (-0.481) + 2.982] = \frac{1}{4} [2.244] = 0.561
$

##### Gradient for Bias $ b $:

$
\frac{\partial \text{Loss}}{\partial b} = \frac{1}{4} [(0.562 - 0) + (0.731 - 1) + (0.679 - 1) + (0.852 - 0)]
$

$
\frac{\partial \text{Loss}}{\partial b} = \frac{1}{4} [0.562 - 0.269 - 0.321 + 0.852] = \frac{1}{4} [0.824] = 0.206
$

#### Step 4: Update the Parameters
Using the gradients, we update the weights and bias using gradient descent:

$
w := w - \alpha \frac{\partial \text{Loss}}{\partial w} = 0.5 - 0.1 \times 0.561 = 0.4439
$

$
b := b - \alpha \frac{\partial \text{Loss}}{\partial b} = 0 - 0.1 \times 0.206 = -0.0206
$

#### Step 5: Summary of Updates
- **Updated weight**: $ w \approx 0.4439 $
- **Updated bias**: $ b \approx -0.0206 $

### Conclusion:
In this one iteration of gradient descent, we:
1. Computed the predictions using the logistic regression formula.
2. Calculated the binary cross-entropy loss.
3. Found the gradients with respect to the weights and bias.
4. Updated the parameters using gradient descent.

This process would be repeated for multiple iterations until the loss converges to a minimum, improving the model's fit to the data.

````python
# Importing modules
from sklearn.datasets import load_iris
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import train_test_split

# Loading dataset
dataset = load_iris()
X = dataset.data
y = dataset.target

# Splitting iris dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 13)

# Creating model
model = PassiveAggressiveClassifier(C = 0.5, random_state = 5)

# Fitting model
model.fit(X_train, y_train)

# Making prediction on test set
test_pred = model.predict(X_test)

# Model evaluation
print(f"Test Set Accuracy : {accuracy_score(y_test, test_pred) * 100} %\n\n")

print(f"Classification Report : \n\n{classification_report(y_test, test_pred)}")
````

In Support Vector Machines (SVMs), the learning objective is to find a hyperplane that maximizes the **margin** between two classes while minimizing classification errors. The margin is the distance between the separating hyperplane and the closest data points (support vectors) of each class.

### Understanding $ \|w\|^{-2} $:

The key idea is to **maximize the margin** between the two classes, which is equivalent to minimizing the norm of the weight vector $ w $. Here's why:

1. **Hyperplane Equation**:
   The hyperplane that separates the two classes is defined by the equation:
   $
   w \cdot x + b = 0
   $
   where $ w $ is the weight vector, $ x $ is the input vector, and $ b $ is the bias term.

2. **Margin**:
   The margin between the two classes is defined as:
   $
   \frac{2}{\|w\|}
   $
   Here, $ \|w\| $ is the Euclidean norm of the weight vector, which measures the magnitude of $ w $. The larger the norm $ \|w\| $, the smaller the margin. Hence, to maximize the margin, we need to minimize $ \|w\| $.

3. **Objective of SVM**:
   SVM aims to maximize this margin (i.e., maximize $ \frac{2}{\|w\|} $), which is equivalent to **minimizing** $ \|w\| $. In mathematical terms, this is often expressed as minimizing $ \frac{1}{2} \|w\|^2 $, where the factor $ \frac{1}{2} $ is included for convenience when taking derivatives during optimization.

### Why $ \|w\|^{-2} $?

- **Geometric Intuition**:
  The margin is proportional to $ \|w\|^{-1} $. The objective is to maximize this margin, which corresponds to minimizing the norm of $ w $. To directly link the margin to the objective function, minimizing $ \|w\|^2 $ (or maximizing $ \|w\|^{-2} $) ensures that the distance between the separating hyperplane and the support vectors is maximized.

- **Relationship to Optimization**:
  SVM optimization problems typically minimize $ \frac{1}{2} \|w\|^2 $, subject to the constraints that all data points are classified correctly or with minimal error (soft-margin SVM). The term $ \|w\|^{-2} $ directly reflects that the goal is to find a hyperplane with the largest margin (smallest norm).

### Key Points:
- $ \|w\|^{-2} $ reflects the margin between the two classes. A smaller $ \|w\| $ means a larger margin, and maximizing the margin helps the SVM generalize better.
- Minimizing $ \|w\|^2 $ (or equivalently maximizing $ \|w\|^{-2} $) is crucial for finding the optimal hyperplane that separates the classes with the maximum possible margin.

In summary, the term $ \|w\|^{-2} $ emphasizes that the margin is inversely proportional to the norm of the weight vector $ w $, and maximizing the margin is the central learning objective of SVM.

Let's walk through a manual example of how to calculate $ \|w\|^{-2} $ using a dummy dataset for SVM. We'll follow these steps:

### Example Setup:

1. **Dummy Dataset**: Suppose we have two linearly separable classes in 2D space:
   - Class 1 points: $ (1, 2) $, $ (2, 3) $, $ (3, 3) $
   - Class 2 points: $ (-1, -1) $, $ (-2, -1) $, $ (-3, -2) $

2. **SVM Objective**: We want to find the optimal hyperplane that separates these two classes and compute the weight vector $ w $ for this hyperplane.

### Step 1: Find the Hyperplane

The general equation of a hyperplane in 2D is:

$
w_1 x_1 + w_2 x_2 + b = 0
$

Where:
- $ w = [w_1, w_2] $ is the weight vector.
- $ b $ is the bias.

Let's assume that we have already found the optimal hyperplane with:
$
w = [2, -1]
$
and $ b = 1 $.

### Step 2: Compute the Norm of $ w $

To find the norm $ \|w\| $, we use the formula:

$
\|w\| = \sqrt{w_1^2 + w_2^2}
$

For our example:

$
\|w\| = \sqrt{2^2 + (-1)^2} = \sqrt{4 + 1} = \sqrt{5} \approx 2.236
$

### Step 3: Compute $ \|w\|^{-2} $

The learning objective in SVM is to maximize the margin, which is proportional to $ \|w\|^{-1} $. So, minimizing $ \|w\|^2 $ (equivalent to maximizing $ \|w\|^{-2} $) gives us the largest margin.

$
\|w\|^2 = 5
$
$
\|w\|^{-2} = \frac{1}{\|w\|^2} = \frac{1}{5} = 0.2
$

### Step 4: Interpretation

- **$ \|w\|^2 $**: This is the value that the SVM tries to minimize, as smaller $ \|w\| $ corresponds to a larger margin.
- **$ \|w\|^{-2} $**: This value (0.2) represents the inverse of the squared norm of $ w $. A larger value of $ \|w\|^{-2} $ indicates a wider margin, which is the goal in SVM.

### Step 5: Margin

The margin is calculated as:

$
\text{Margin} = \frac{2}{\|w\|} = \frac{2}{2.236} \approx 0.894
$

This margin represents the distance between the separating hyperplane and the closest data points (support vectors).

### Conclusion:

In this manual example:
- The weight vector $ w = [2, -1] $.
- The norm $ \|w\| \approx 2.236 $.
- The squared norm $ \|w\|^2 = 5 $.
- The inverse squared norm $ \|w\|^{-2} = 0.2 $.

The margin between the classes is $ 0.894 $, and the SVM optimization goal is to maximize this margin by minimizing $ \|w\|^2 $.

Let's walk through how the weight vector $ w $ is updated in a simple Support Vector Machine (SVM) during the training process using a linear SVM. We'll focus on a couple of iterations to manually understand how the weights are adjusted.

### Key Concepts:

- In SVM, the objective is to find a hyperplane that maximizes the margin between two classes.
- The hyperplane is defined by $ w \cdot x + b = 0 $, where $ w $ is the weight vector, $ x $ is the input, and $ b $ is the bias.
- SVM updates the weights using a gradient-based optimization algorithm. The hinge loss function is used to measure how far the predictions are from the true labels.
  
The weight update rule is:

$
w_{t+1} = w_t - \eta \frac{\partial L}{\partial w_t}
$

where $ \eta $ is the learning rate, $ L $ is the hinge loss, and $ w_t $ is the current weight vector.

### Simple Example Setup:

Let's consider a simple 2D example with the following data points:

- Class 1: $ (2, 3) $, $ (4, 5) $ labeled as $ y = +1 $
- Class 2: $ (-2, -1) $, $ (-3, -4) $ labeled as $ y = -1 $

Assume the initial weight vector is:
$
w = [0, 0], \quad b = 0
$
The learning rate $ \eta = 0.1 $.

### Step-by-Step Update for a Couple of Iterations:

#### Iteration 1:

1. **Data Point**: $ (2, 3) $, $ y = +1 $

   - **Prediction**: Using $ w \cdot x + b = 0 \cdot 2 + 0 \cdot 3 + 0 = 0 $, the prediction is 0.
   - **Hinge Loss**: Since the prediction is incorrect (for $ y = +1 $), the hinge loss is:
     $
     L = \max(0, 1 - y (w \cdot x + b)) = \max(0, 1 - (1)(0)) = 1
     $
   - **Weight Update**: We update the weight vector using the rule:
     $
     w_{new} = w_{old} + \eta y x
     $
     Substituting values:
     $
     w_{new} = [0, 0] + 0.1 \times 1 \times [2, 3] = [0.2, 0.3]
     $
     So, after this step, the weight vector becomes $ w = [0.2, 0.3] $.

#### Iteration 2:

2. **Data Point**: $ (-2, -1) $, $ y = -1 $

   - **Prediction**: Using the updated weight vector $ w \cdot x + b = 0.2 \cdot (-2) + 0.3 \cdot (-1) = -0.4 - 0.3 = -0.7 $, the prediction is negative, which matches $ y = -1 $, but we check the margin.
   - **Hinge Loss**: The hinge loss is:
     $
     L = \max(0, 1 - y (w \cdot x + b)) = \max(0, 1 - (-1)(-0.7)) = \max(0, 1 - 0.7) = 0.3
     $
   - **Weight Update**: Since there is a small hinge loss, we update the weight vector:
     $
     w_{new} = [0.2, 0.3] + 0.1 \times (-1) \times [-2, -1] = [0.2, 0.3] + [0.2, 0.1] = [0.4, 0.4]
     $
     So, after this step, the weight vector becomes $ w = [0.4, 0.4] $.

#### Iteration 3:

3. **Data Point**: $ (4, 5) $, $ y = +1 $

   - **Prediction**: Using the new weight vector $ w \cdot x + b = 0.4 \cdot 4 + 0.4 \cdot 5 = 1.6 + 2.0 = 3.6 $, the prediction is positive, which is correct for $ y = +1 $.
   - **Hinge Loss**: Since the prediction is positive and sufficiently large, the hinge loss is:
     $
     L = \max(0, 1 - y (w \cdot x + b)) = \max(0, 1 - (1)(3.6)) = \max(0, 1 - 3.6) = 0
     $
   - **No Update**: Since the loss is 0, no update is needed, and the weight vector remains $ w = [0.4, 0.4] $.

### Summary of First Few Iterations:

- After the first data point $ (2, 3) $, $ w = [0.2, 0.3] $.
- After the second data point $ (-2, -1) $, $ w = [0.4, 0.4] $.
- After the third data point $ (4, 5) $, the weight vector did not change.

In each iteration, the SVM adjusts the weights in response to the classification errors. If the classification is correct and the margin is large enough, the weights remain the same. If the classification is incorrect or the margin is too small, the weights are updated in the direction that reduces the error, effectively pushing the hyperplane to better separate the classes.

This process continues until the model converges to the optimal hyperplane.

The **kernel trick** in Support Vector Machines (SVM) is a mathematical technique that allows SVMs to solve non-linearly separable problems by implicitly mapping the data into a higher-dimensional space, without explicitly computing the coordinates in that space. Here’s an intuitive breakdown:

### 1. **Linear SVM**:
In simple SVMs, the algorithm finds a hyperplane that best separates data points into different classes. This works well when the data is linearly separable, meaning that a straight line (in 2D) or a flat plane (in higher dimensions) can separate the classes.

### 2. **Non-linear Problems**:
However, real-world data is often not linearly separable. For example, in two-dimensional space, the classes might form concentric circles or have complex shapes that cannot be separated by a straight line.

### 3. **Mapping to Higher Dimensions**:
One way to handle this is to transform the data into a higher-dimensional space where it becomes linearly separable. For instance, by mapping 2D data points to 3D, the SVM might find a hyperplane that separates the data in this new space. Mathematically, this transformation is done using a **feature map** function $\phi(x)$.

### 4. **The Problem with High-dimensional Mapping**:
While this approach can work, calculating the actual coordinates of the data in this higher-dimensional space can be computationally expensive, especially when dealing with very high-dimensional spaces.

### 5. **The Kernel Trick**:
This is where the kernel trick comes in. Instead of explicitly calculating the transformation and the dot products in the higher-dimensional space, the **kernel trick** uses a special function, called a **kernel function**, to compute the dot product in the transformed space directly, without ever needing to know the coordinates of the points in that space.

In other words, instead of computing $\phi(x_i)$ and $\phi(x_j)$ and then taking their dot product $\phi(x_i) \cdot \phi(x_j)$, the kernel trick allows you to compute this dot product using a kernel function $K(x_i, x_j)$ in the original space.

### 6. **Common Kernel Functions**:
Some popular kernel functions include:
- **Linear kernel**: $ K(x_i, x_j) = x_i \cdot x_j $
- **Polynomial kernel**: $ K(x_i, x_j) = (x_i \cdot x_j + c)^d $, where $c$ is a constant and $d$ is the degree of the polynomial.
- **Radial Basis Function (RBF) or Gaussian kernel**: $ K(x_i, x_j) = \exp\left(-\frac{\|x_i - x_j\|^2}{2\sigma^2}\right) $, where $\sigma$ controls the width of the Gaussian.

Let's reconsider the **toy example** with non-linearly separable data and show how the **kernel trick** works mathematically using a small dataset. We'll choose an example where the data is clearly non-linearly separable in its original 2D form.

### Toy Example Setup:
We have two classes of points arranged in a circular pattern:

- Class 1 (label $ y = 1 $): $ (1, 0), (0, 1) $
- Class 2 (label $ y = -1 $): $ (-1, 0), (0, -1) $

In 2D space, these points cannot be separated by a straight line. We’ll use the **kernel trick** with a **polynomial kernel** to handle this.

### Step 1: Polynomial Kernel
The **polynomial kernel** is a popular choice for mapping data into higher dimensions. The polynomial kernel of degree 2 is defined as:

$
K(x_i, x_j) = (x_i \cdot x_j + 1)^2
$

This kernel corresponds to implicitly mapping the points to a higher-dimensional space where the features are products of the original features.

### Step 2: Compute the Kernel Matrix
Let’s calculate the kernel for the four points. We’ll compute the kernel matrix, which contains all the dot products between pairs of points after applying the kernel trick.

#### Data Points:
- $ x_1 = (1, 0) $
- $ x_2 = (0, 1) $
- $ x_3 = (-1, 0) $
- $ x_4 = (0, -1) $

We’ll compute the kernel for each pair:

$
K(x_1, x_1) = (x_1 \cdot x_1 + 1)^2 = ((1 \cdot 1) + (0 \cdot 0) + 1)^2 = (1 + 1)^2 = 4
$
$
K(x_1, x_2) = (x_1 \cdot x_2 + 1)^2 = ((1 \cdot 0) + (0 \cdot 1) + 1)^2 = (0 + 1)^2 = 1
$
$
K(x_1, x_3) = (x_1 \cdot x_3 + 1)^2 = ((1 \cdot -1) + (0 \cdot 0) + 1)^2 = (-1 + 1)^2 = 0
$
$
K(x_1, x_4) = (x_1 \cdot x_4 + 1)^2 = ((1 \cdot 0) + (0 \cdot -1) + 1)^2 = (0 + 1)^2 = 1
$

Now, let’s compute the rest similarly:

$
K(x_2, x_2) = (x_2 \cdot x_2 + 1)^2 = ((0 \cdot 0) + (1 \cdot 1) + 1)^2 = (1 + 1)^2 = 4
$
$
K(x_2, x_3) = (x_2 \cdot x_3 + 1)^2 = ((0 \cdot -1) + (1 \cdot 0) + 1)^2 = (0 + 1)^2 = 1
$
$
K(x_2, x_4) = (x_2 \cdot x_4 + 1)^2 = ((0 \cdot 0) + (1 \cdot -1) + 1)^2 = (-1 + 1)^2 = 0
$
$
K(x_3, x_3) = (x_3 \cdot x_3 + 1)^2 = ((-1 \cdot -1) + (0 \cdot 0) + 1)^2 = (1 + 1)^2 = 4
$
$
K(x_3, x_4) = (x_3 \cdot x_4 + 1)^2 = ((-1 \cdot 0) + (0 \cdot -1) + 1)^2 = (0 + 1)^2 = 1
$
$
K(x_4, x_4) = (x_4 \cdot x_4 + 1)^2 = ((0 \cdot 0) + (-1 \cdot -1) + 1)^2 = (1 + 1)^2 = 4
$

### Step 3: Kernel Matrix
Now, we can form the **kernel matrix** $ K $:

$
K = \begin{pmatrix}
4 & 1 & 0 & 1 \\
1 & 4 & 1 & 0 \\
0 & 1 & 4 & 1 \\
1 & 0 & 1 & 4
\end{pmatrix}
$

This matrix represents the dot products between the data points as if they had been mapped to a higher-dimensional space. Using this matrix, the **SVM** can now find a separating hyperplane in this new feature space, even though we never explicitly computed the coordinates of the points in that space.

### Step 4: Why This Works
The kernel trick avoids explicitly transforming the data into a higher-dimensional space. Instead, the **kernel function** computes the dot product of points as though they were mapped to the higher space, allowing the SVM to work with non-linearly separable data.

### Size of the Kernel Matrix

The **kernel matrix** (also called the **Gram matrix**) will indeed grow larger as the number of data points increases. Here's why:

The kernel matrix $K$ is an $n \times n$ matrix, where $n$ is the number of data points. Each entry $ K(x_i, x_j) $ in the matrix represents the kernel value (or dot product in the transformed space) between data points $x_i$ and $x_j$.

Thus:
- If you have $n$ data points, the kernel matrix will have $n^2$ entries.
- For small datasets, this isn't a big issue, but for large datasets, the kernel matrix becomes quite large and can consume a lot of memory.

For example:
- **100 data points** → kernel matrix of size $ 100 \times 100 $ (10,000 entries).
- **1,000 data points** → kernel matrix of size $ 1,000 \times 1,000 $ (1 million entries).
- **10,000 data points** → kernel matrix of size $ 10,000 \times 10,000 $ (100 million entries).

### Computational Cost
- **Memory usage**: The kernel matrix has $ n^2 $ entries, which can be problematic for large datasets. It requires storing all these values, which could lead to memory issues.
- **Time complexity**: Calculating each entry of the kernel matrix (i.e., computing the kernel function between every pair of data points) takes time. For example, in a **Radial Basis Function (RBF)** kernel, calculating the exponential of the distance between points can be time-consuming if you have to do it for every pair of points.

### Practical Implications for Large Datasets
When dealing with large datasets, the kernel matrix's size becomes a bottleneck for **memory** and **computational power**. Some ways to mitigate this issue include:

1. **Approximation techniques**: Techniques like the **Nyström method** or **random Fourier features** approximate the kernel matrix to reduce computational complexity.
2. **Linear SVMs**: For very large datasets, linear SVMs (without a kernel) may work reasonably well, and they scale much better than kernel-based SVMs.
3. **Distributed SVMs**: Break the dataset into smaller batches and distribute the computation across multiple machines.

### Summary:
- The kernel matrix grows quadratically with the number of data points.
- For large datasets, the kernel matrix can become too large to handle, requiring significant memory and computation.
- There are methods to approximate or reduce the computational load of working with large kernel matrices.



### 7. **Benefits**:
- **Efficiency**: The kernel trick allows SVMs to perform complex separations in high-dimensional spaces without ever needing to explicitly compute those dimensions.
- **Flexibility**: By choosing different kernels, SVMs can handle a wide variety of data distributions and separations.

### Example:
Imagine a set of points that form two concentric circles (a classic non-linear problem). A **linear SVM** would fail to separate these points in 2D. However, by applying a **kernel trick** with an appropriate kernel (like the RBF kernel), the SVM can implicitly map these points into a higher-dimensional space, where they become linearly separable.

# XGBOOST

### XGBoost (Extreme Gradient Boosting) - Short Notes

**XGBoost** is a powerful machine learning algorithm based on gradient boosting that is designed for speed and performance. It has gained popularity due to its efficiency and accuracy in supervised learning tasks like classification and regression.

#### Key Concepts:

1. **Boosting**:
   - **Boosting** is an ensemble learning technique where models are trained sequentially, and each model corrects the errors of its predecessor.
   - In **gradient boosting**, each new model is trained to minimize the errors (residuals) of the previous models using gradient descent.

2. **Decision Trees**:
   - XGBoost uses decision trees as base learners. Instead of building one large tree, it builds several small trees that focus on improving performance by correcting the previous trees' errors.

3. **Objective**:
   - XGBoost optimizes a regularized objective function:
     $
     \text{Objective} = \text{Loss Function} + \text{Regularization Term}
     $
   - The **loss function** measures how well the model fits the training data (e.g., mean squared error for regression or log loss for classification).
   - The **regularization term** penalizes model complexity to avoid overfitting.

4. **Regularization**:
   - XGBoost includes **L1** (Lasso) and **L2** (Ridge) regularization to control the complexity of the model.
   - This helps prevent overfitting by discouraging overly complex models.

5. **Gradient Descent**:
   - XGBoost improves models iteratively using gradient descent, where each new tree is trained to minimize the gradient of the loss with respect to the predictions.

6. **Tree Pruning**:
   - XGBoost uses a technique called **max_depth** to control the depth of the trees, which limits overfitting by preventing deep trees that may over-learn the training data.

7. **Handling Missing Values**:
   - XGBoost can automatically handle missing data by learning the best direction to take in the decision tree for missing values.

8. **Feature Importance**:
   - XGBoost provides ways to measure **feature importance**, helping identify the most relevant features in the data.

#### Advantages:

1. **Speed**: XGBoost is highly optimized for speed, using parallel processing and optimized tree building algorithms.
2. **Regularization**: It includes built-in regularization, which helps prevent overfitting and ensures better generalization.
3. **Handling Imbalanced Data**: XGBoost has hyperparameters like `scale_pos_weight` that help balance imbalanced datasets.
4. **Sparsity Awareness**: It can handle sparse data efficiently, especially useful for datasets with missing values or sparse feature vectors.
5. **Cross-validation**: XGBoost supports efficient cross-validation through its in-built functions.

#### Hyperparameters:
- **Learning Rate ($\eta$)**: Controls the size of the steps taken to minimize the loss. Lower values slow down learning but often result in better models.
- **Max Depth**: Controls the depth of each tree. Larger depths may lead to overfitting.
- **n_estimators**: Number of trees (iterations) to build.
- **subsample**: Proportion of data used to train each tree (used to prevent overfitting).
- **colsample_bytree**: Proportion of features used to build each tree.

#### Use Cases:
- Classification tasks (binary and multi-class)
- Regression tasks
- Ranking problems
- Time series forecasting

#### Applications:
- Kaggle competitions (XGBoost is widely used)
- Finance (e.g., credit scoring)
- Healthcare (e.g., patient risk prediction)
- Recommendation systems

XGBoost is a highly flexible, efficient, and scalable algorithm suitable for both small datasets and large-scale problems. It is widely used in machine learning competitions and industry for its accuracy and ability to handle a wide variety of data structures.

### Handling Missing Values in XGBoost:

XGBoost can **handle missing values automatically**. When XGBoost encounters a missing value during training or prediction, it learns the best direction to take in the decision tree. This process works as follows:

- During the tree-building process, XGBoost treats missing values as a separate condition.
- It finds the optimal direction (left or right in the tree) based on the loss reduction achieved when routing missing values in either direction.
- In practice, this allows XGBoost to make robust predictions even when data has missing entries, without the need for manual imputation.

### Impact of Outliers on XGBoost:

Outliers can have an impact on XGBoost because they influence the **gradients** calculated during the boosting process. However, XGBoost is generally robust to outliers due to the following reasons:

1. **Robust Loss Functions**: You can choose loss functions that are less sensitive to outliers, such as Huber loss for regression, which reduces the effect of extreme values.
2. **Ensemble Nature**: Since XGBoost is based on boosting trees, it mitigates the effect of individual outliers by aggregating predictions across multiple trees.
3. **Regularization**: L1 (Lasso) and L2 (Ridge) regularization in XGBoost helps to reduce the impact of noisy data (outliers) by penalizing large coefficients.

**Tip**: If outliers are a concern, you can apply preprocessing techniques like **clipping** or **removing outliers** before feeding the data into XGBoost.

### Application Areas of XGBoost:

1. **Classification**:
   - Binary classification (e.g., spam detection, fraud detection).
   - Multi-class classification (e.g., categorizing news articles, image classification).
   
2. **Regression**:
   - Predicting continuous variables (e.g., housing prices, stock market trends).

3. **Ranking**:
   - Applications like **information retrieval** (e.g., search engine results ranking).
   - Personalized recommendation systems.

4. **Time Series Forecasting**:
   - Forecasting future values based on historical data (e.g., sales forecasting).

5. **Anomaly Detection**:
   - Detecting unusual patterns in the data (e.g., fraud detection, network intrusion detection).

### Tips and Tricks for Using XGBoost:

1. **Tuning Hyperparameters**:
   - **Learning Rate ($\eta$)**: Start with a low learning rate (e.g., 0.01 or 0.1) and increase the number of trees (`n_estimators`).
   - **Max Depth**: Shallower trees (e.g., max depth = 3 or 4) are usually better to avoid overfitting.
   - **Regularization**: Use **L1** or **L2 regularization** to prevent overfitting.
   - **Subsampling**: Set `subsample` (percentage of rows used per tree) and `colsample_bytree` (percentage of features used) to improve model generalization.

2. **Early Stopping**:
   - Use **early stopping** to prevent overfitting. Stop training when the validation loss stops improving.

3. **Cross-validation**:
   - Utilize XGBoost’s built-in **cross-validation** function (`xgb.cv()`) to automatically determine the best hyperparameters.

4. **Scale-Pos-Weight**:
   - For imbalanced classification, set `scale_pos_weight` to balance positive and negative classes.

5. **Handling Missing Values**:
   - XGBoost automatically handles missing values, so no need to manually impute them.

6. **Parallelism**:
   - XGBoost supports **parallel training** and GPU acceleration, making it efficient on large datasets.

# Manual Computation

Let’s walk through a **manual example** of how XGBoost works with **dummy data** without using code. We will break the steps down to show how the data is processed, how the decision trees are built, and how the model makes predictions.

### Dummy Data Setup:

Suppose we have the following **dummy dataset** for a binary classification problem:

| Instance | Feature 1 (X1) | Feature 2 (X2) | Label (y) |
|----------|----------------|----------------|-----------|
| 1        | 2              | 3              | 1         |
| 2        | 1              | 2              | 1         |
| 3        | 4              | 5              | 0         |
| 4        | 6              | 7              | 0         |
| 5        | 3              | 4              | 1         |

### Objective:

Our goal is to predict the labels (1 or 0) for the instances using the XGBoost algorithm. We'll do the following:

1. **Step 1: Initialize the prediction to a base value** (typically the mean of the labels).
2. **Step 2: Compute the residuals** (difference between actual labels and predictions).
3. **Step 3: Build decision trees** based on residuals to reduce the error.
4. **Step 4: Update the predictions** based on the output of the trees.
5. **Step 5: Iterate** this process until the model converges.

---

### Step 1: Initialize Prediction

- **Initial prediction**: XGBoost starts by making an initial guess for all instances. The simplest guess is the mean of the target variable $ y $.

- **Mean of labels**: The labels $ y $ are $[1, 1, 0, 0, 1]$. The mean is:

$
\text{Initial Prediction} = \frac{1 + 1 + 0 + 0 + 1}{5} = 0.6
$

So, for all instances, we initially predict $ y = 0.6 $.

### Step 2: Compute Residuals

Residuals measure how far the initial predictions are from the actual labels. The formula for residuals is:

$
\text{Residual} = \text{Actual Label} - \text{Prediction}
$

| Instance | Actual Label (y) | Initial Prediction | Residual |
|----------|------------------|--------------------|----------|
| 1        | 1                | 0.6                | 0.4      |
| 2        | 1                | 0.6                | 0.4      |
| 3        | 0                | 0.6                | -0.6     |
| 4        | 0                | 0.6                | -0.6     |
| 5        | 1                | 0.6                | 0.4      |

### Step 3: Build the First Tree

Next, XGBoost builds a **decision tree** that tries to predict the **residuals**. Let’s construct a simple decision tree based on the features $ X1 $ and $ X2 $:

1. **Split** on **Feature 1 (X1)** at a threshold of 3:
   - If $ X1 \leq 3 $, go to the left.
   - If $ X1 > 3 $, go to the right.

This split gives the following groups:
- **Left Node**: Instances 1, 2, and 5 (where $ X1 \leq 3 $)
- **Right Node**: Instances 3 and 4 (where $ X1 > 3 $)

### Step 4: Assign Output Values to Each Node

For each node, we compute the **output value** based on the average of the residuals in that node:

- **Left Node** (Instances 1, 2, and 5):
  - Residuals: $ 0.4, 0.4, 0.4 $
  - Output: $ \frac{0.4 + 0.4 + 0.4}{3} = 0.4 $

- **Right Node** (Instances 3 and 4):
  - Residuals: $ -0.6, -0.6 $
  - Output: $ \frac{-0.6 + -0.6}{2} = -0.6 $

### Step 5: Update the Predictions

The updated predictions are based on the initial prediction plus a **learning rate** multiplied by the output of the decision tree. Assume a **learning rate** of 0.1.

- **Updated Prediction**:
  $
  \text{New Prediction} = \text{Old Prediction} + 0.1 \times \text{Tree Output}
  $

| Instance | Old Prediction | Tree Output | Updated Prediction |
|----------|----------------|-------------|--------------------|
| 1        | 0.6            | 0.4         | $ 0.6 + 0.1 \times 0.4 = 0.64 $ |
| 2        | 0.6            | 0.4         | $ 0.6 + 0.1 \times 0.4 = 0.64 $ |
| 3        | 0.6            | -0.6        | $ 0.6 + 0.1 \times (-0.6) = 0.54 $ |
| 4        | 0.6            | -0.6        | $ 0.6 + 0.1 \times (-0.6) = 0.54 $ |
| 5        | 0.6            | 0.4         | $ 0.6 + 0.1 \times 0.4 = 0.64 $ |

### Step 6: Iterate

XGBoost repeats this process:
1. Compute new residuals based on the updated predictions.
2. Build a new tree to predict the residuals.
3. Update the predictions.

The model continues to refine the predictions by building additional trees. With each tree, the model becomes more accurate at predicting the actual labels.

### Conclusion

In this dummy example:
1. XGBoost starts with an initial prediction (mean of labels).
2. It computes residuals (errors) and builds a decision tree to minimize these errors.
3. The tree outputs corrections, and the predictions are updated.
4. This process iterates until the model converges or a stopping criterion is met.

XGBoost uses **boosting** to sequentially correct the errors of previous trees, making it powerful for structured/tabular data and both classification and regression tasks.
